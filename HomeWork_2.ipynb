### Задание 2


```python
t = np.linspace(0, 10, 51)
f = np.cos(t)

plt.title("График f(t)")
plt.xlabel("Значения t", color = "purple")
plt.ylabel("Значения f", color = "purple")

plt.plot(t, f, color = "green", label = "cos(t)")

legend = plt.legend(loc = "lower right")
plt.setp(legend.get_texts(), color = "Blue")

plt.axis([0.5, 9.5, -2.5, 2.5])


plt.show()
```
